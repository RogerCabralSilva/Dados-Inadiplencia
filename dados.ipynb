previsores e alvo

````python
y = dados['inadimplente']
x = dados.drop(columns="inadimplente")
````

oversmapling

````python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)

x, y = sm.fit_resample(x, y)

print(y.value_counts())
````

Divisão de dados

````python
from sklearn.model_selection import train_test_split
````

````python
X_treino, X_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.20, random_state=42, stratify=y)
````

# Modelos

Random Tree

````python
from sklearn.ensemble import RandomForestClassifier

modelo_random_forest = RandomForestClassifier(max_depth=12)

modelo_random_forest.fit(X_treino, y_treino)

y_predit = modelo_random_forest.predict(X_teste)
````

Avaliações

````python
from sklearn.metrics import classification_report, RocCurveDisplay, ConfusionMatrixDisplay
````

````python
print(classification_report(y_teste, y_predit))
````

````python
RocCurveDisplay.from_predictions(y_teste, y_predit)
````

````python
ConfusionMatrixDisplay.from_predictions(y_teste, y_predit)
````

KNN 

````python
from imblearn.pipeline import Pipeline as impipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
````

````python
pipeline_knn = impipeline([("StanderScaler", StandardScaler()), ("KNN", KNeighborsClassifier())])

pipeline_knn.fit(X_treino, y_treino)

y_predit = pipeline_knn.predict(X_teste)
````

````python
print(classification_report(y_teste, y_predit))
````

````python
ConfusionMatrixDisplay.from_predictions(y_teste, y_predit)
````

````python
RocCurveDisplay.from_predictions(y_teste, y_predit)
````

SVM

````python
from sklearn.svm import SVC
````

````python
pipeline_svc = impipeline([("StanderScaler", StandardScaler()), ("svc", SVC())])

pipeline_svc.fit(X_treino, y_treino)

y_predit = pipeline_svc.predict(X_teste)
````

````python
print(classification_report(y_teste, y_predit))
````

````python
ConfusionMatrixDisplay.from_predictions(y_teste, y_predit)
````

````python
RocCurveDisplay.from_predictions(y_teste, y_predit)
````

Redes Neurais

````python
from sklearn.neural_network import MLPClassifier
````

````python
pipeline_MPL = impipeline([("StanderScaler", StandardScaler()), ("svc", MLPClassifier(max_iter=1000))])

pipeline_MPL.fit(X_treino, y_treino)

y_predit = pipeline_MPL.predict(X_teste)
````

````python
print(classification_report(y_teste, y_predit))
````

````python
ConfusionMatrixDisplay.from_predictions(y_teste, y_predit)
````

````python
RocCurveDisplay.from_predictions(y_teste, y_predit)
````

XGboost

````python
import xgboost as xgb
````

````python
xg_cl = xgb.XGBClassifier(objective='binary:logistic')
xg_cl.fit(X_treino, y_treino)

previsao = xg_cl.predict(X_teste)

print(classification_report(y_teste, previsao))
ConfusionMatrixDisplay.from_predictions(y_teste, previsao)
RocCurveDisplay.from_predictions(y_teste, previsao)
````

# Combinação de modelos

Melhores Parâmetros

````python
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import GridSearchCV, cross_validate, StratifiedKFold
````

Random Forest

````python
parametros_random_forest = {
  "n_estimators" : [10, 25, 50, 100],
  "criterion" : ["gini", "entropy", "log_loss"],
  "max_depth" : [10, 20, 30, 40],
  #"min_samples_split" : [2, 5, 10],
  #"min_samples_leaf" : [1, 2, 4],
  #"max_features" : ["auto", "sqrt"]
  
}
````

````python
grid_rf = GridSearchCV(estimator=modelo_random_forest,
                       param_grid=parametros_random_forest,
                       n_jobs=-1,
                       cv=5).fit(X_treino, y_treino)
````

````python
grid_rf.best_params_
````

````python
skf = StratifiedKFold(shuffle=True, random_state=0)

cv = cross_validate(grid_rf, x, y, cv=skf, scoring="recall")
````

````python
cv
````

Melhores Parâmetros

XGboost

````python
xgb_params = {
    'booster': ["gbtree", "gblinear", "dart"],  # Tipo de booster: 'gbtree', 'gblinear', ou 'dart'
    'learning_rate': [0.01, 0.001, 1],  # Taxa de aprendizado (eta), controla o impacto de cada árvore
    'max_depth': [6, 10, 20],  # Profundidade máxima da árvore
}
````

````python
grid_xgb = GridSearchCV(estimator=xg_cl, 
                        param_grid=xgb_params,
                        cv=skf,
                        n_jobs=-1,
                        ).fit(X_treino, y_treino)
````

````python
grid_xgb.best_params_
````

Voting

````python
modelo_voting = VotingClassifier(
  estimators=[
    ("rf", RandomForestClassifier(criterion='log_loss', max_depth=30, n_estimators=100)), ("xgb", xg_cl)
  ],
  voting='soft'
)

modelo_voting.fit(X_treino, y_treino)

y_predit = modelo_voting.predict(X_teste)
````

````python
print(classification_report(y_teste, y_predit))
ConfusionMatrixDisplay.from_predictions(y_teste, y_predit)
RocCurveDisplay.from_predictions(y_teste, y_predit)
````

validação cruzada

````python
cv = cross_validate(modelo_voting, x, y, cv=skf, scoring='recall')

cv
````
